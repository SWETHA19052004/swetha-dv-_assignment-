The code installs the required packages (geopandas and folium) using the !pip install commands.
Import Libraries:

Import the necessary libraries (geopandas, matplotlib.pyplot, folium).
Load GeoJSON Data:

Reads GeoJSON data from the file located at '/content/export.geojson' using geopandas.read_file() and stores it in the GeoDataFrame (gdf).
Plotting with Geopandas:

Uses matplotlib.pyplot to create a scatter plot of hospitals and supermarkets from the GeoDataFrame (gdf). Hospitals are marked in red (H marker) and supermarkets in blue (s marker).
Adds legend, title, and labels to the plot.
Show the Plot:

Displays the plot using plt.show().
Create Folium Map:

Creates a Folium map centered at the mean latitude and longitude of the GeoDataFrame.
Add Markers to Folium Map:

Iterates through hospitals and supermarkets in the GeoDataFrame and adds markers to the Folium map. Hospitals are marked in red, and supermarkets in blue. The marker popups display the names of the corresponding locations.
Show the Folium Map:

Displays the Folium map.

```python
# Install necessary packages
!pip install geopandas
!pip install folium

import geopandas as gpd
import matplotlib.pyplot as plt
import folium

# Load GeoJSON data
file_path = '/content/export.geojson'
gdf = gpd.read_file("/content/export.geojson")

# Plot the data using geopandas
fig, ax = plt.subplots(figsize=(10, 10))
gdf[gdf['amenity'] == 'hospital'].plot(ax=ax, color='red', marker='H', markersize=50, label='Hospitals')
gdf[gdf['shop'] == 'supermarket'].plot(ax=ax, color='blue', marker='s', markersize=50, label='Supermarkets')

# Add legend and labels
ax.legend()
ax.set_title('Distribution of Hospitals and Supermarkets')
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')

# Show the plot
plt.show()

# Create a folium map
m = folium.Map(location=[gdf.geometry.y.mean(), gdf.geometry.x.mean()], zoom_start=12)

# Add hospitals as red markers
for idx, hospital in gdf[gdf['amenity'] == 'hospital'].iterrows():
    folium.Marker([hospital.geometry.y, hospital.geometry.x], icon=folium.Icon(color='red'), popup=hospital['name']).add_to(m)

# Add supermarkets as blue markers
for idx, supermarket in gdf[gdf['shop'] == 'supermarket'].iterrows():
    folium.Marker([supermarket.geometry.y, supermarket.geometry.x], icon=folium.Icon(color='blue'), popup=supermarket['name']).add_to(m)

# Show the map
m
```

Map is plotted with grocery shops and Hospitals with some different icons

The below code, designed for execution in a Colab environment, utilizes the matplotlib library to create visualizations comparing the distribution of hospitals and grocery stores in a GeoDataFrame (gdf). After filtering the GeoDataFrame for hospitals and grocery stores, the code calculates the counts for each amenity type. It then generates a side-by-side display with a pie chart on the left, illustrating the percentage distribution of hospitals and grocery stores, and a bar chart on the right, providing a direct comparison of their counts. The pie chart uses colors red and blue for hospitals and grocery stores, respectively, and both charts are labeled with informative titles. The resulting visualizations offer insights into the spatial distribution and relative abundance of hospitals and grocery stores within the geographical data.

```python
import matplotlib.pyplot as plt

# Filter the GeoDataFrame for hospitals and grocery stores
hospitals = gdf[gdf['amenity'] == 'hospital']
grocery_stores = gdf[gdf['shop'] == 'supermarket']

# Count the number of each amenity type
hospital_counts = hospitals.shape[0]
grocery_counts = grocery_stores.shape[0]

# Plot a pie chart
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Pie chart
axes[0].pie([hospital_counts, grocery_counts], labels=['Hospitals', 'Grocery Stores'], autopct='%1.1f%%', startangle=90)
axes[0].axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
axes[0].set_title('Distribution of Hospitals and Grocery Stores')

# Bar chart
amenity_types = ['Hospitals', 'Grocery Stores']
counts = [hospital_counts, grocery_counts]

axes[1].bar(amenity_types, counts, color=['red', 'blue'])
axes[1].set_ylabel('Count')
axes[1].set_title('Comparison of Hospitals and Grocery Stores')

plt.show()
```

This Python code, using the matplotlib library, creates visualizations to analyze the distribution of amenities, particularly hospitals and grocery stores, within a GeoDataFrame (gdf). The code begins by counting the occurrences of each amenity type and then filters the GeoDataFrame to separate hospitals and grocery stores. Two sets of visualizations are generated: the first set consists of two pie charts displayed side by side. The left chart illustrates the overall distribution of amenities, presenting the percentage share for each amenity type in the dataset. The right chart focuses specifically on hospitals and grocery stores, showing their relative proportions within the filtered data. The second set of visualizations is a bar chart displaying the counts of hospitals and grocery stores. The bars are colored red for hospitals and blue for grocery stores, offering a clear visual comparison. These visualizations provide insights into the spatial distribution and prevalence of different amenities, aiding in the analysis of geographical patterns.

```python
import matplotlib.pyplot as plt

# Count the number of hospitals and grocery stores
amenity_counts = gdf['amenity'].value_counts()

# Filter the data for hospitals and grocery stores
hospitals = gdf[gdf['amenity'] == 'hospital']
grocery_stores = gdf[gdf['shop'] == 'supermarket']

# Count the number of hospitals and grocery stores
hospital_count = len(hospitals)
grocery_count = len(grocery_stores)

# Plot a pie chart
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

# Pie chart for overall amenity distribution
ax[0].pie(amenity_counts, labels=amenity_counts.index, autopct='%1.1f%%', startangle=90)
ax[0].axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
ax[0].set_title('Distribution of Amenities')

# Pie chart for hospitals and grocery stores
ax[1].pie([hospital_count, grocery_count], labels=['Hospitals', 'Grocery Stores'], autopct='%1.1f%%', startangle=90)
ax[1].axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
ax[1].set_title('Distribution of Hospitals and Grocery Stores')

plt.show()

# Bar chart
fig, ax = plt.subplots(figsize=(8, 6))

# Bar chart for hospitals and grocery stores
ax.bar(['Hospitals', 'Grocery Stores'], [hospital_count, grocery_count], color=['red', 'blue'])
ax.set_title('Distribution of Hospitals and Grocery Stores')
ax.set_ylabel('Count')

plt.show()
```
