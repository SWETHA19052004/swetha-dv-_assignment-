```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset
data = {
    'Roll no': ['10DS03001', '10DS03002', '10DS03003', '10DS03004', '10DS03005', '10DS03006', '10DS03007', '10DS03008', '10DS03009', '10DS03010'],
    'Attendance': [64, 93, 85, 70, 66, 67, 64, 68, 68, 91],
    'Grades': [440, 537, 464, 416, 484, 542, 523, 428, 441, 452]
}

df = pd.DataFrame(data)

# Visualize the distribution of grades using box and violin plots
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.boxplot(df['Grades'])
plt.title('Distribution of Grades (Box Plot)')

plt.subplot(1, 2, 2)
parts = plt.violinplot(df['Grades'], showmedians=True)
for pc in parts['bodies']:
    pc.set_facecolor('lightblue')
    pc.set_edgecolor('black')
plt.title('Distribution of Grades (Violin Plot)')

plt.tight_layout()
plt.show()

# Explore the relationship between attendance and grades using a scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(df['Attendance'], df['Grades'])
plt.title('Relationship between Attendance and Grades')
plt.xlabel('Attendance')
plt.ylabel('Grades')
plt.grid(True)
plt.show()
```
