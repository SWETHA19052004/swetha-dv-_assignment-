```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create a DataFrame from the provided dataset
data = pd.read_csv('/content/Order Details.csv')

# Let's first identify the most popular products using a bar chart.
plt.figure(figsize=(12, 6))
sns.barplot(x='Amount', y='Sub-Category', data=data, estimator=sum, ci=None)
plt.title('Most Popular Products by Sales')
plt.xlabel('Total Sales Amount')
plt.ylabel('Sub-Category')
plt.show()

# Now, let's identify the most popular categories using a pie chart.
category_counts = data['Category'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Sales by Category')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
